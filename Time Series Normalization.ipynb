```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
biggles=pd.read_csv('TimeTraces.csv',header=None)
```

```python
reet=pd.read_csv('Treatments.csv',header=None)
```

```python
biggles.head()
```

```python
for i in range(len(biggles)):
    mean=biggles.iloc[i].mean()
    std=biggles.iloc[i].std()
    for j in range(600):
        biggles.at[i,j]=((biggles.at[i,j]-mean)/std)
```

```python
biggles.head()
```

```python
biggles["treat"]=reet
```

```python
nibbles=biggles.sample(1000)
```

```python
for i in range(1000):
    if nibbles.iloc[i][600]==0:
        plt.plot(nibbles.iloc[i].index[:600],nibbles.iloc[i][:600],alpha=.01,color="r")
    elif nibbles.iloc[i][600]==1:
        plt.plot(nibbles.iloc[i].index[:600],nibbles.iloc[i][:600],alpha=.01,color="g")
    else:
        plt.plot(nibbles.iloc[i].index[:600],nibbles.iloc[i][:600],alpha=.01,color="b")
```
